```python
import pandas as pd
import numpy as np

from pycaret.regression import setup,compare_models
from lightgbm import LGBMRegressor
```

```python
train=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\UD\\Regression with abalone dataset\\train.csv")
test=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\UD\\Regression with abalone dataset\\test.csv")
```

```python
train
```

```python
test
```

```python
target=train.Rings
test_id=test.id

train_1=train.drop(['id','Rings'],axis=1)
test_1=test.drop('id',axis=1)

data=pd.concat([train_1,test_1],axis=0).reset_index(drop=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
data=data.drop('Sex_M',axis=1)
```

```python
data
```

```python
train_final=data.loc[:train.index.max(),:].copy()
test_final=data.loc[train.index.max()+1:,:].reset_index(drop=True).copy()
```

```python
_=setup(data=pd.concat([train_final,target],axis=1),target='Rings')
```

```python
compare_models()
```

```python
lgbm=LGBMRegressor()

lgbm.fit(train_final,target)
```

```python
pred=lgbm.predict(test_final)
```

```python
pred
```

```python
pred=np.round_(pred)
```

```python
pred
```

```python
pred=pred.astype('int')
```

```python
pred
```

```python
final_pred=pd.concat([test_id,pd.Series(pred,name='Rings')],axis=1)
```

```python
final_pred
```
